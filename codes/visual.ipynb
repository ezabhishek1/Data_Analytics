```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

data = sns.load_dataset('iris')
print(data.head())
```

```python
sns.pairplot(data, hue='species', diag_kind='kde',palette='Set1')
plt.show()
```

```python
sns.jointplot(data=data,x='sepal_length', y='sepal_width',  kind='scatter', hue='species')
plt.show()
```

```python
sns.jointplot(
    data=data,
    x='sepal_length',
    y='sepal_width',
    kind='hex'
)

plt.show()
```

```python
sns.jointplot(
    data=data,
    x='sepal_length',
    y='sepal_width',
    kind='reg'
)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset('iris')

plt.figure(figsize=(10, 8))

# Select only numeric columns and compute correlation matrix
corr_matrix = iris.select_dtypes(include='number').corr()

# Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',
            linewidths=3, linecolor='white')

plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
```

```python
import plotly.express as px
#interactive scatterplot
fig=px.scatter(iris, x="sepal_length", y="sepal_width",color="species", size="petal_length", hover_data=["petal_width"] )
fig.show()
```
