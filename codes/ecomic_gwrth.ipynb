```python
import pandas as pd

df = pd.read_csv(r"D:\testing\BCT_3\data\india_economic_growth_large_dataset.csv")

df.head()
```

```python
import seaborn as sns
sns.histplot(df["GDP_Growth_%"])
```

```python
df["Leading_Sector"].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
import matplotlib.pyplot as plt
yearly = df.groupby("Year")["GDP_Growth_%"].mean()
plt.plot(yearly.index, yearly.values)
plt.title("India GDP Growth Trend (Future Prediction)")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")

plt.title("Economic Indicators Correlation")

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df[["FDI_Billion_$","Tech_Investment_Billion_$","Exports_Billion_$"]]
y = df["GDP_Growth_%"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

prediction = model.predict(X_test)

print("Sample Predictions:")
print(prediction[:5])
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, prediction)

print("Model Accuracy:", score)
```

```python
sns.scatterplot(x="FDI_Billion_$", y="GDP_Growth_%", data=df)

plt.title("FDI vs GDP Growth")

plt.show()
```
