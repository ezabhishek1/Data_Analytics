```python
import pandas as pd

# Load dataset
df = pd.read_excel(r"D:\testing\BCT_3\data\social_media_trends_dataset.xlsx")

# View first rows
print(df.head())

# Average Likes per Post
avg_likes = df["Likes"].mean()
print("Average Likes per Post:", avg_likes)

# Average Comments per Post
avg_comments = df["Comments"].mean()
print("Average Comments per Post:", avg_comments)

# Total number of posts
total_posts = len(df)

# Engagement Ratio
engagement_ratio = (df["Likes"].sum() + df["Comments"].sum() + df["Shares"].sum()) / total_posts
print("Engagement Ratio:", engagement_ratio)

# Create Engagement column
df["Engagement"] = df["Likes"] + df["Comments"] + df["Shares"]

# Find post with highest engagement
highest_post = df.loc[df["Engagement"].idxmax()]
print("Post with Highest Engagement:")
print(highest_post)
```
