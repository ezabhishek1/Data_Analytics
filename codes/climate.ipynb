```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate last 20 years
years = np.arange(2004, 2024)

# Simulated average temperature data for Kolkata
np.random.seed(10)
temperature = 26 + np.linspace(0,1.5,20) + np.random.normal(0,0.2,20)

# Create dataframe
data = pd.DataFrame({
    "Year": years,
    "Avg_Temperature_C": temperature
})

print(data)

# Plot the trend
plt.figure(figsize=(10,6))
plt.plot(data["Year"], data["Avg_Temperature_C"], marker='o', linewidth=2)

plt.title("Average Temperature Change in Kolkata (Last 20 Years)")
plt.xlabel("Year")
plt.ylabel("Average Temperature (°C)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

years = list(range(2004, 2024))

temperature = [
26.1, 26.3, 26.2, 26.4, 26.5,
26.6, 26.7, 26.8, 26.9, 27.0,
27.1, 27.2, 27.3, 27.4, 27.5,
27.6, 27.7, 27.8, 27.9, 28.0
]

data = pd.DataFrame({
    "Year": years,
    "Temperature": temperature
})

# Create category
data["Temp_Level"] = ["Low" if t < 26.8 else "High" for t in data["Temperature"]]

# Pairplot
sns.pairplot(data, hue="Temp_Level", diag_kind="kde", palette="Set2")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data
years = list(range(2004, 2024))

temperature = [
26.1, 26.3, 26.2, 26.4, 26.5,
26.6, 26.7, 26.8, 26.9, 27.0,
27.1, 27.2, 27.3, 27.4, 27.5,
27.6, 27.7, 27.8, 27.9, 28.0
]

data = pd.DataFrame({
    "Year": years,
    "Temperature": temperature
})

# Create category for hue
data["Temp_Level"] = ["Low" if t < 26.8 else "High" for t in data["Temperature"]]

# Pairplot with histogram
sns.pairplot(data, hue="Temp_Level", diag_kind="hist", palette="Set2")
plt.show()

# Histplot
plt.figure(figsize=(8,5))
sns.histplot(data["Temperature"], kde=True, bins=8)

plt.title("Temperature Distribution (Kolkata)")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Climate dataset (example)
data = pd.DataFrame({
    "Year": list(range(2004, 2024)),
    "Temperature": [
        26.1, 26.3, 26.2, 26.4, 26.5,
        26.6, 26.7, 26.8, 26.9, 27.0,
        27.1, 27.2, 27.3, 27.4, 27.5,
        27.6, 27.7, 27.8, 27.9, 28.0
    ],
    "Rainfall": [
        1500,1480,1510,1520,1490,
        1475,1505,1495,1510,1485,
        1470,1490,1500,1480,1475,
        1460,1485,1470,1465,1450
    ],
    "Humidity": [
        70,71,69,72,73,
        74,72,75,76,74,
        73,75,77,76,78,
        79,78,80,81,82
    ]
})

# Correlation matrix
corr_matrix = data.corr()

# Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',
            linewidths=3, linecolor='white')

plt.title("Correlation Matrix of Climate Dataset")

plt.show()
```
